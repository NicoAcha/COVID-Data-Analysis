```python
columnas = [
    "country_name",
    "date",
    "new_confirmed",
    "cumulative_confirmed",
    "cumulative_vaccine_doses_administered",
    "population",
    "population_density",
    "human_development_index"
]

df_reducido = df[columnas].copy()
```

## Para aportar mayor significación al análisis, nos centramos en columnas de datos referentes a contexto (que no varían significativamente en el timpo y describen plenamente al país) y no a variables de eventos, por tal motivo, analizaremos los nulos de las variables population, population_density y human_development_index.- 

```python
df_reducido[["population","human_development_index","population_density"]].isna().sum()
```

## Realizamos un relevamiento de nulos y comprobamos que population posee aprox. un 4% de nulos en todos sus registros (lo que permite imputar y llenar datos por país), pero que human_development_index posee mas del 50% de nulos y que population_density casi todos sus registros nulos, lo cual nos termina dando certeza que el fill lo debemos hacer sobre valores del mismo país. Comprendiendo que utilizaremos primero forward fill (hacia adelante) y luego backward fill (si aún quedan nulos, rellenando hacia atrás).-

```python
nulos_fill = ["population","human_development_index","population_density"]
```

```python
df_reducido[nulos_fill] = df_reducido.groupby("country_name")[nulos_fill].ffill()
```

```python
df_reducido[nulos_fill] = df_reducido.groupby("country_name")[nulos_fill].bfill()
```

```python
df_pais_sorted = df_pais.sort_values("date")
df_ultimo_valor = df_pais_sorted.groupby("country_name").last().reset_index()
```

## Volvemos a crear otro df ordenado por fecha (de mas antigua a mas nueva), y lo asignamos dentro de un nuevo df donde agrupando por pais verificamos el último registro válido pero al hacer esto, "country_name" pasa a ser índice, y por lo cual, se lo reinicia mediante reset_index() logrando así una nueva tabla con datos certeros.-

## Habiendo realizado todas las verificaciones necesarias, para llegar a los datos sólidos, procedemos al comienzo del análisis basándonos en que para tratar la expansión de laboratorios, es mas relevante evaluar los casos por millón que población total, ya que eso impacta en la presión sobre el sistema de salud, y la necesidad relativa de diagnóstico y control.-

```python
df_analisis = df_ultimo_valor[columnas_analisis].copy()
```

```python
df_analisis["cases_per_million"] = (
    df_analisis["cumulative_confirmed"] / df_analisis["population"] * 1000000
)
```

## Se crea dentro de df_analisis la variable de "cases_per_million" que será el resultado de dividir los casos confirmados por la población y multiplicarlo por 1.000.000 para evaluar la incidencia relativa y no solo valores absolutos (al normalizar los casos acumulados por millón de habitantes)

```python
df_analisis.sort_values("cases_per_million",ascending=False).round(0)
```

## Insight: Para evaluar el impacto proporcional de la pandemia en cada país, se normalizaron los casos por millón de habitantes. El ranking muestra países como Uruguay, Chile, Panamá, Costa y Argentina con una mayor carga relativa de enfermedad, lo que indica una mayor presión sanitaria en relación al tamaño poblacional. Esto nos permite priorizar regiones donde necesitamos infraestructura de diagnóstico y seguimiento, ya que podría determinar un punto crítico mas allá de los valores absolutos de casos. 

## ¿Por qué se excluye Puerto Rico que lidera el primer puesto del ranking? Porque presenta alta incidencia pero información incompleta (valores NaN), y por lo cual se prioriza el análisis en países con datos completos.-

```python
df_analisis["doses_per_one_hundred"] = (
    df_analisis["cumulative_vaccine_doses_administered"] / df_analisis["population"] * 100
)
```

## Obtenemos una tabla que nos indique la proporción de dosis cada 100 habitantes para relativizar el impacto de la población en la vacunación.-

```python
df_analisis.sort_values("doses_per_one_hundred",ascending=False).round(0)
```

```python
df_analisis.sort_values("cases_per_million",ascending=False)[
    ["country_name","cases_per_million","doses_per_one_hundred"]
].round(0)
```

## Cruzamos ambos datos (casos por millón, vacunas por cien) para obtener una métrica mas certera y visualizar a continuación donde debemos poner el foco.-

```python
inc_med = df_analisis["cases_per_million"].median()
vac_med = df_analisis["doses_per_one_hundred"].median()

df_prioridad = df_analisis [
    (df_analisis["cases_per_million"] > inc_med) &
    (df_analisis["doses_per_one_hundred"] < vac_med)
]

df_prioridad
```

## Se logra detectar aquellos países que cuentan con una alta incidencia pero con menor cobertura de vacunación, lo que los posiciona como regiones prioritarias para fortalecer la infraestructura de diagnóstico y control. Siendo candidatos principales, para la expansión de laboratorios.-

```python
df_prioridad.sort_values("population_density",ascending=False) [
    ["country_name","cases_per_million","doses_per_one_hundred","population_density"]
].round(1)
```

## El análisis multivariable indica que Colombia y Brasil representan prioridad absoluta para la expansión de laboratorios, al combinar alta incidencia relativa (casos por millón de habitantes), cobertura de vacunación moderada (dosis por cada 100 habitantes) y condiciones de densidad poblacional que favorecen la propagación del virus. 

## Paraguay, si bien se ubica dentro del grupo de alta incidencia y baja vacunación, presenta menor densidad poblacional, lo que sugiere un menor riesgo de transmisión masiva. Por este motivo, se posiciona como prioridad secundaria, orientada principalmente al fortalecimiento de la capacidad sanitaria más que a la expansión de alto volumen de diagnóstico.

```python
columnas_numericas = [
    "cases_per_million",
    "doses_per_one_hundred",
    "population_density"
]
```

```python
for col in columnas_numericas:
    media = df_analisis[col].mean()
    minimo = df_analisis[col].min()
    maximo = df_analisis[col].max()

    print(f"\nEstadisticas para {col}:")
    print(f"Media: {media:.2f}")
    print(f"Minimo: {minimo:.2f}")
    print(f"Maximo: {maximo:.2f}")
```

## Las métricas calculadas permiten comprender la distribución de los datos. La media representa el valor promedio de cada indicador, mientras que los valores mínimo y máximo muestran los extremos del conjunto. Estas métricas ayudan a identificar la magnitud típica de los indicadores y posibles valores atípicos.

## No todas las columnas del dataset son aptas para este tipo de análisis, ya que variables categóricas como el nombre del país no representan valores cuantitativos. Analizar columnas incorrectas podría llevar a interpretaciones erróneas de los resultados.

```python
def medidas(col):
    mediana = col.median()
    varianza = col.var()
    rango = col.max() - col.min()

    return mediana, varianza, rango
```

```python
for col in columnas_numericas:
    med, var, ran = medidas(df_analisis[col])
    print(col)
    print(f"Mediana: {med:.2f}, Varianza: {var:.2f}, Rango: {ran:.2f}")
    
```

## A raíz de las medidas anteriores podemos notar que en los casos por millon, tenemos una varianza altísima, lo que implica países muy distintos. En torno a dosis cada 100 personas, observamos una varianza mucho menor, logrando objetivizar politicas mas homogenas de vacunación. Y por último en la densidad poblacional observamos una dispersion grande que hace alusión a geografías muy distintas.

```python
df_pais[["country_name", "date", "new_confirmed", "cumulative_confirmed"]].head()
```

```python
df_pais["month"] = df_pais["date"].dt.to_period("M")
```

## Incorporamos la columna "month" para realizar una progresión mas certera sobre el crecimiento / decrecimiento de los casos, tomando la fecha relativa y convirtiéndola en datos de meses.-

```python
df_pais[["country_name","month","new_confirmed","cumulative_confirmed"]].head()
```

```python
casos_mensuales = (
    df_pais
    .groupby(["country_name","month"])["new_confirmed"]
    .sum()
    .reset_index()
)

casos_mensuales
```

```python
paises_claves = ["Cuba","Chile"]

df_plot = casos_mensuales[casos_mensuales["country_name"].isin(paises_claves)]
df_plot = df_plot.sort_values("month")
```

```python
plt.figure(figsize=(10,6))

for pais in paises_claves:
    data_pais = df_plot[df_plot["country_name"] == pais]
    plt.plot(data_pais["month"].astype(str),data_pais["new_confirmed"],label=pais)

plt.xticks(rotation=45)
plt.title("Evolución mensual de nuevos casos - Cuba / Chile")
plt.ylabel("Casos nuevos - Suma mensual")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
picos = df_plot.loc[
    df_plot.groupby("country_name")["new_confirmed"].idxmax()
]

picos[["country_name","month","new_confirmed"]]
```

## Mediante este gráfico, se termina de confirmar lo que veníamos analizando. Se observa que los picos de contagios no ocurren simultáneamente entre países. Cuba presenta su mayor pico de casos en Agosto del 2021, mientras que Chile lo hace varios meses después, en Febrero del 2022.

## Esta desincronización sugiere que la demana de infraestructura sanitaria no se concentra en los mismos períodos en toda la región, por lo que la planificación de expansión de laboratorios debe considerar estrategias escalonadas en el tiempo y no solo focalizadas en la región.-

```python
paises_objetivo = ["Colombia","Brazil","Paraguay"]
```

```python
df_objetivo = casos_mensuales[casos_mensuales["country_name"].isin(paises_objetivo)]
df_objetivo = df_objetivo.sort_values("month")
```

```python
plt.figure(figsize=(10,6))

for pais in paises_objetivo:
    data_pais = df_objetivo[df_objetivo["country_name"] == pais]
    plt.plot(data_pais["month"].astype(str), data_pais["new_confirmed"], label=pais)

plt.xticks(rotation=45)
plt.title("Evolución mensual de casos nuevos – Países prioritarios")
plt.xlabel("Mes")
plt.ylabel("Casos nuevos por mes")
plt.legend()
plt.tight_layout()
plt.show()
```

## Evolución temporal en países prioritarios

## Para la variable de dosis por cada 100 habitantes fue necesario utilizar funciones que ignoran valores faltantes (NaN), lo que evidencia la importancia de un tratamiento adecuado de datos incompletos durante el análisis exploratorio.

```python
x = df_analisis["doses_per_one_hundred"].values
y = df_analisis["cases_per_million"].values

mask = ~np.isnan(x) & ~np.isnan(y)

corr_np = np.corrcoef(x[mask], y[mask])[0,1]
corr_np
```
